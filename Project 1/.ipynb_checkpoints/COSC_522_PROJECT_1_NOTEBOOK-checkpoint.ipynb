```python
X = df.drop("Label", axis =1)
y = df.Label
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# List to store accuracy scores for different values of k
k_values = range(1, 50)
accuracies = []

# Loop over different values of k
for k in k_values:
    # Initialize KNN with current k
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Fit the model on the training data
    knn.fit(X_train, y_train)
    
    # Predict on the testing set
    y_pred = knn.predict(X_test)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Plot k vs accuracy
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o', linestyle='-')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('KNN Accuracy vs. Number of Neighbors (k)')
plt.grid(True)
plt.xticks(k_values)
plt.show()
```

## Implementation of Baye's Classification Algorithm

```python
# Set random seed for reproducibility
np.random.seed(42)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Function to calculate class-wise accuracy
def class_wise_accuracy(y_true, y_pred):
    report = classification_report(y_true, y_pred, output_dict=True)
    class_0_acc = report['0']['recall']
    class_1_acc = report['1']['recall']
    return class_0_acc, class_1_acc

# Initialize results storage
results = []

# Naive Bayes
start_time = time.time()
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred_gnb = gnb.predict(X_test)
overall_acc_gnb = accuracy_score(y_test, y_pred_gnb)
class_0_acc_gnb, class_1_acc_gnb = class_wise_accuracy(y_test, y_pred_gnb)
runtime_gnb = time.time() - start_time
results.append(['Naive Bayes', overall_acc_gnb, class_0_acc_gnb, class_1_acc_gnb, runtime_gnb])

# K-Nearest Neighbors (KNN)
start_time = time.time()
knn = KNeighborsClassifier(n_neighbors=5)  # Using k=5 as a default value
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
overall_acc_knn = accuracy_score(y_test, y_pred_knn)
class_0_acc_knn, class_1_acc_knn = class_wise_accuracy(y_test, y_pred_knn)
runtime_knn = time.time() - start_time
results.append(['K-Nearest Neighbors', overall_acc_knn, class_0_acc_knn, class_1_acc_knn, runtime_knn])

# Parzen-Rosenblatt Window (Kernel Density Estimation)
# Implementing using KDE and predicting based on density
start_time = time.time()
# Fit KDE for each class separately
kde_class_0 = KernelDensity(bandwidth=1.0).fit(X_train[y_train == 0])
kde_class_1 = KernelDensity(bandwidth=1.0).fit(X_train[y_train == 1])

# Predict class based on which density is higher
log_density_0 = kde_class_0.score_samples(X_test)
log_density_1 = kde_class_1.score_samples(X_test)
y_pred_kde = np.where(log_density_0 > log_density_1, 0, 1)

overall_acc_kde = accuracy_score(y_test, y_pred_kde)
class_0_acc_kde, class_1_acc_kde = class_wise_accuracy(y_test, y_pred_kde)
runtime_kde = time.time() - start_time
results.append(['Parzen-Rosenblatt Window', overall_acc_kde, class_0_acc_kde, class_1_acc_kde, runtime_kde])

# Create a DataFrame for results
results_df = pd.DataFrame(results, columns=['Algorithm', 'Overall Accuracy', 'Class 0 Accuracy', 'Class 1 Accuracy', 'Runtime (seconds)'])

# Display results
print(results_df)
```

```python
from matplotlib.colors import ListedColormap

def plot_decision_boundary(clf, X, y, ax, title):
    # Create meshgrid for plotting decision boundaries
    x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
    y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))
    
    # Predict on the meshgrid
    if hasattr(clf, "predict"):
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    else:  # Special handling for Parzen-Rosenblatt Window
        log_density_0 = kde_class_0.score_samples(np.c_[xx.ravel(), yy.ravel()])
        log_density_1 = kde_class_1.score_samples(np.c_[xx.ravel(), yy.ravel()])
        Z = np.where(log_density_0 > log_density_1, 0, 1).reshape(xx.shape)
    
    # Plot the decision boundary and scatter plot the data points
    ax.contourf(xx, yy, Z, alpha=0.3, cmap=ListedColormap(['#FFAAAA', '#AAAAFF']))
    
    # Correct indexing to use .loc instead of .iloc for boolean conditions
    ax.scatter(X.loc[y == 0, X.columns[0]], X.loc[y == 0, X.columns[1]], c='red', label='Class 0', edgecolors='k')
    ax.scatter(X.loc[y == 1, X.columns[0]], X.loc[y == 1, X.columns[1]], c='blue', label='Class 1', edgecolors='k')
    
    ax.set_title(title)
    ax.legend()

# Plotting decision boundaries
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

plot_decision_boundary(gnb, X, y, axes[0], 'Naive Bayes Decision Boundary')
plot_decision_boundary(knn, X, y, axes[1], 'KNN Decision Boundary (k=5)')
plot_decision_boundary(None, X, y, axes[2], 'Parzen-Rosenblatt Decision Boundary')

plt.tight_layout()
plt.show()
```
