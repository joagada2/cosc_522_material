# TASK TWO
### Implementing KNN Without Using Any Python Framework using Euclidean Distance as Preferred Distance Measure

```python
# concatenate the previously generated dataset into a training set and a test set
df_training = pd.concat([df_1_training, df_2_training, df_3_training], ignore_index=True)
df_test = pd.concat([df_1_test, df_2_test, df_3_test], ignore_index=True)
```

```python

# Split data into training and testing sets without using the tran_test_split module of Sklearn
X_train = df_training[['Feature1', 'Feature2']].values
y_train = df_training['Label'].values

X_test = df_test[['Feature1', 'Feature2']].values
y_test = df_test['Label'].values

# Function to calculate Euclidean distance
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

# KNN function without using any frameworks
def knn_predict(X_train, y_train, X_test, k):
    predictions = []
    for test_point in X_test:
        # Calculate distances from the test point to all training points
        distances = [euclidean_distance(test_point, x) for x in X_train]
        # Get the indices of the k nearest neighbors
        k_indices = np.argsort(distances)[:k]
        # Get the labels of the k nearest neighbors
        k_nearest_labels = [y_train[i] for i in k_indices]
        # Determine the most common label
        most_common = Counter(k_nearest_labels).most_common(1)
        predictions.append(most_common[0][0])
    return np.array(predictions)

# Function to calculate accuracy
def calculate_accuracy(y_true, y_pred):
    return np.sum(y_true == y_pred) / len(y_true)

# Testing different values of k
k_values = range(1, 50)
accuracies = []

for k in k_values:
    y_pred = knn_predict(X_train, y_train, X_test, k)
    accuracy = calculate_accuracy(y_test, y_pred)
    accuracies.append(accuracy)
    print(f"Accuracy for k={k}: {accuracy:.2f}")

# Plotting k values against their accuracies
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o', linestyle='-')
plt.title('KNN Accuracy vs. k Value')
plt.xlabel('k Value')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# KNN Classifier Implementation with k=3
def knn_classifier(X_train, y_train, X_test, k=3):
    predictions = []
    for x in X_test:
        # Calculate Euclidean distances to all training samples
        distances = np.linalg.norm(X_train - x, axis=1)
        # Get the indices of the k nearest neighbors
        k_indices = np.argsort(distances)[:k]
        # Predict the class with the majority vote
        k_nearest_labels = y_train[k_indices]
        prediction = np.argmax(np.bincount(k_nearest_labels.astype(int)))
        predictions.append(prediction)
    return np.array(predictions)

# Quadratic Discriminant Classifier Implementation
def qda_classifier(X_train, y_train, X_test):
    mean_1 = np.mean(X_train[y_train == 1], axis=0)
    mean_2 = np.mean(X_train[y_train == 0], axis=0)
    cov_1 = np.cov(X_train[y_train == 1], rowvar=False)
    cov_2 = np.cov(X_train[y_train == 0], rowvar=False)
    predictions = []
    for x in X_test:
        # Compute discriminant function for both classes
        g1 = -0.5 * np.dot(np.dot((x - mean_1), np.linalg.inv(cov_1)), (x - mean_1).T) - 0.5 * np.log(np.linalg.det(cov_1))
        g2 = -0.5 * np.dot(np.dot((x - mean_2), np.linalg.inv(cov_2)), (x - mean_2).T) - 0.5 * np.log(np.linalg.det(cov_2))
        predictions.append(1 if g1 > g2 else 0)
    return np.array(predictions)

# Mahalanobis Distance Classifier Implementation
def mahalanobis_classifier(X_train, y_train, X_test):
    mean_1 = np.mean(X_train[y_train == 1], axis=0)
    mean_2 = np.mean(X_train[y_train == 0], axis=0)
    cov = np.cov(X_train, rowvar=False)
    inv_cov = np.linalg.inv(cov)
    predictions = []
    for x in X_test:
        dist_to_1 = np.sqrt(np.dot(np.dot((x - mean_1), inv_cov), (x - mean_1).T))
        dist_to_2 = np.sqrt(np.dot(np.dot((x - mean_2), inv_cov), (x - mean_2).T))
        predictions.append(1 if dist_to_1 < dist_to_2 else 0)
    return np.array(predictions)

# Euclidean Distance Classifier Implementation
def euclidean_classifier(X_train, y_train, X_test):
    mean_1 = np.mean(X_train[y_train == 1], axis=0)
    mean_2 = np.mean(X_train[y_train == 0], axis=0)
    predictions = []
    for x in X_test:
        dist_to_1 = np.linalg.norm(x - mean_1)
        dist_to_2 = np.linalg.norm(x - mean_2)
        predictions.append(1 if dist_to_1 < dist_to_2 else 0)
    return np.array(predictions)

# Function to calculate accuracy and classwise accuracy
def calculate_accuracy(y_true, y_pred):
    overall_accuracy = np.mean(y_true == y_pred)
    class_1_accuracy = np.mean(y_pred[y_true == 1] == 1)
    class_2_accuracy = np.mean(y_pred[y_true == 0] == 0)
    return overall_accuracy, class_1_accuracy, class_2_accuracy

# Collecting results for all classifiers
results = []

# KNN Classifier
start_time = time.time()
y_pred_knn = knn_classifier(X_train, y_train, X_test, k=3)
knn_overall, knn_class1, knn_class2 = calculate_accuracy(y_test, y_pred_knn)
knn_time = time.time() - start_time
results.append(['KNN', knn_overall, knn_class1, knn_class2, knn_time])

# QDA Classifier
start_time = time.time()
y_pred_qda = qda_classifier(X_train, y_train, X_test)
qda_overall, qda_class1, qda_class2 = calculate_accuracy(y_test, y_pred_qda)
qda_time = time.time() - start_time
results.append(['QDA', qda_overall, qda_class1, qda_class2, qda_time])

# Mahalanobis Distance Classifier
start_time = time.time()
y_pred_mahalanobis = mahalanobis_classifier(X_train, y_train, X_test)
maha_overall, maha_class1, maha_class2 = calculate_accuracy(y_test, y_pred_mahalanobis)
maha_time = time.time() - start_time
results.append(['Mahalanobis', maha_overall, maha_class1, maha_class2, maha_time])

# Euclidean Distance Classifier
start_time = time.time()
y_pred_euclidean = euclidean_classifier(X_train, y_train, X_test)
euclidean_overall, euclidean_class1, euclidean_class2 = calculate_accuracy(y_test, y_pred_euclidean)
euclidean_time = time.time() - start_time
results.append(['Euclidean', euclidean_overall, euclidean_class1, euclidean_class2, euclidean_time])

# Printing results in tabular format
print("Summary of Classification Performance:")
print(f"{'Classifier':<15}{'Overall Accuracy':<20}{'Class 1 Accuracy':<20}{'Class 2 Accuracy':<20}{'Run Time (s)':<10}")
for row in results:
    print(f"{row[0]:<15}{row[1]:<20.4f}{row[2]:<20.4f}{row[3]:<20.4f}{row[4]:<10.4f}")
```

```python
# Visualization of decision boundaries
def plot_decision_boundaries(X_test, y_test, classifier_func, title):
    # Plotting the decision boundary using meshgrid
    x_min, x_max = min(x[0] for x in X_test) - 1, max(x[0] for x in X_train) + 1
    y_min, y_max = min(x[1] for x in X_test) - 1, max(x[1] for x in X_train) + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    Z = np.array([classifier_func(X_train, y_train, [[xi, yi]])[0] for xi, yi in zip(np.ravel(xx), np.ravel(yy))])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(*zip(*X_train), c=y_train, edgecolor='k', s=50)
    plt.title(title)

plt.figure(figsize=(12, 10))
plt.subplot(2, 2, 1)
plot_decision_boundaries(X_train, y_train, knn_classifier, "k-NN (k=20)")

plt.subplot(2, 2, 2)
plot_decision_boundaries(X_train, y_train, euclidean_classifier, "Euclidean Classifier")

plt.subplot(2, 2, 3)
plot_decision_boundaries(X_train, y_train, mahalanobis_classifier, "Mahalanobis Classifier")

plt.subplot(2, 2, 4)
plot_decision_boundaries(X_train, y_train, qda_classifier, "QDA Classifier")

plt.tight_layout()
plt.show()
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Parameters for Gaussian distributions
mean_w1 = [2, 2]           # Mean for class 1
cov_w1 = [[1, 0.5], [0.5, 1]]   # Covariance matrix for class 1

mean_w2 = [-2, -2]         # Mean for class 2
cov_w2 = [[1, -0.5], [-0.5, 1]] # Covariance matrix for class 2

# Generating training set: 300 samples from class 1, 100 samples from class 2
X_train_w1 = np.random.multivariate_normal(mean_w1, cov_w1, 300)
y_train_w1 = np.ones(300)  # Label for class 1

X_train_w2 = np.random.multivariate_normal(mean_w2, cov_w2, 100)
y_train_w2 = np.zeros(100) # Label for class 2

X_train = np.vstack((X_train_w1, X_train_w2))
y_train = np.hstack((y_train_w1, y_train_w2))

# Generating testing set: 150 samples from class 1, 30 samples from class 2
X_test_w1 = np.random.multivariate_normal(mean_w1, cov_w1, 150)
y_test_w1 = np.ones(150)   # Label for class 1

X_test_w2 = np.random.multivariate_normal(mean_w2, cov_w2, 30)
y_test_w2 = np.zeros(30)   # Label for class 2

X_test = np.vstack((X_test_w1, X_test_w2))
y_test = np.hstack((y_test_w1, y_test_w2))

# Minimum Euclidean distance classifier
def euclidean_distance_classifier(X_train, y_train, X_test, prior_w1):
    # Compute class means
    mean_class_1 = np.mean(X_train[y_train == 1], axis=0)
    mean_class_2 = np.mean(X_train[y_train == 0], axis=0)

    # Classify each point in the test set
    predictions = []
    for x in X_test:
        # Calculate distances to class means
        dist_to_w1 = np.linalg.norm(x - mean_class_1)
        dist_to_w2 = np.linalg.norm(x - mean_class_2)
        
        # Bayesian decision rule considering priors
        if dist_to_w1 * (1 / prior_w1) < dist_to_w2 * (1 / (1 - prior_w1)):
            predictions.append(1)  # Classify as class 1
        else:
            predictions.append(0)  # Classify as class 2
            
    return np.array(predictions)

# Calculate classification accuracy for varying prior probabilities of P(w1)
priors = np.linspace(0.01, 0.99, 100)  # Avoid extreme cases of 0 and 1
accuracies = []

for prior_w1 in priors:
    y_pred = euclidean_distance_classifier(X_train, y_train, X_test, prior_w1)
    accuracy = np.mean(y_pred == y_test)  # Calculate accuracy
    accuracies.append(accuracy)

# Plotting the classification accuracy vs. P(w1)
plt.figure(figsize=(10, 6))
plt.plot(priors, accuracies, marker='o', color='b')
plt.title("Classification Accuracy vs. P(w1)")
plt.xlabel("P(w1)")
plt.ylabel("Overall Classification Accuracy")
plt.grid(True)
plt.show()
```
