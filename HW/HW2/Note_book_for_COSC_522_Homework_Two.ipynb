## b (i)
#### Code for implementing gradient descent with starting point at t = 7, leaning rates of 0.01, 0.05, 0.1, and 0.5

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function f(t) and its derivative (gradient)
def cost_function(t):
    return 50 * np.sin(t) + t**2

def gradient(t):
    return 50 * np.cos(t) + 2 * t

# Gradient descent implementation
def gradient_descent(starting_point, learning_rate, num_iterations):
    t = starting_point
    path = [t]  # To store the path of convergence
    
    for _ in range(num_iterations):
        grad = gradient(t)
        t = t - learning_rate * grad
        path.append(t)
        # Stop early if the gradient is very close to zero (converged)
        if abs(grad) < 1e-5:
            break
    
    return t, path

# Settings
starting_point = 7
learning_rates = [0.01, 0.05, 0.1, 0.5]  # Different learning rates
num_iterations = 100  # Maximum number of iterations

# Plot the cost function
t_values = np.linspace(-10, 10, 1000)
f_values = cost_function(t_values)

plt.figure(figsize=(14, 8))
plt.plot(t_values, f_values, label='f(t) = 50*sin(t) + t^2', color='blue')

# Run gradient descent with different learning rates
for lr in learning_rates:
    final_t, path = gradient_descent(starting_point, lr, num_iterations)
    path_values = [cost_function(t) for t in path]
    
    # Plot the path of convergence
    plt.plot(path, path_values, marker='o', label=f'Learning Rate: {lr}, Converged t: {final_t:.2f}')
    print(f"Learning Rate: {lr} | Final t: {final_t:.4f} | Cost: {cost_function(final_t):.4f}")

plt.title('Gradient Descent Convergence Paths')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.legend()
plt.grid(True)
plt.show()
```

# PROBLEM 2 (b)

```python
# Initial parameters
w1, w2 = 0.1, 0.2
b = 0.5
learning_rate = 1
max_iterations = 10

# OR Gate Table
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
targets = [0, 1, 1, 1]

# Function to calculate perceptron output
def perceptron_output(x1, x2, w1, w2, b):
    z = w1 * x1 + w2 * x2 + b
    return 1 if z >= 0 else 0

# Training the perceptron
for epoch in range(max_iterations):
    print(f"Iteration {epoch + 1}")
    for (x1, x2), target in zip(inputs, targets):
        # Calculate the perceptron output
        prediction = perceptron_output(x1, x2, w1, w2, b)
        
        # Display the current status
        print(f"Inputs: ({x1}, {x2}), Target: {target}, Perceptron Output: {prediction}")
        
        # Calculate the error
        error = target - prediction
        
        # Update the weights and bias
        w1 += learning_rate * error * x1
        w2 += learning_rate * error * x2
        b += learning_rate * error
    
    # Display updated weights and bias after each epoch
    print(f"Updated weights: w1 = {w1}, w2 = {w2}, bias = {b}\n")
```
